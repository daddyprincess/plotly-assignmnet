## Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
# Import the necessary libraries
import seaborn as sns
import plotly.express as px

# Load the Titanic dataset using Seaborn's load_dataset function
titanic_data = sns.load_dataset("titanic")

# Create a scatter plot using Plotly Express
scatter_plot = px.scatter(titanic_data, x="age", y="fare", title="Scatter Plot of Age vs Fare in Titanic Dataset")

# Show the plot
scatter_plot.show()
```

## Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
# Import the necessary libraries
import plotly.express as px

# Load the "tips" dataset from Plotly Express
tips_data = px.data.tips

# Create a box plot using Plotly Express
box_plot = px.box(tips_data, x="day", y="total_bill", title="Box Plot of Total Bill by Day")

# Show the plot
box_plot.show()
```

## Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
# Import the necessary libraries
import plotly.express as px

# Load the "tips" dataset from Plotly Express
tips_data = px.data.tips

# Create a histogram with pattern shapes and color mapping
histogram = px.histogram(
    tips_data,
    x="sex",
    y="total_bill",
    color="day",  # Use the "day" column for color
    facet_col="smoker",  # Use the "smoker" column for pattern shapes
    pattern_shape="smoker",  # Set pattern shapes based on the "smoker" column
    title="Histogram of Total Bill by Sex (Colored by Day, Patterned by Smoker)"
)

# Show the plot
histogram.show()
```

## Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

```python
# Import the necessary libraries
import plotly.express as px

# Load the "iris" dataset from Plotly Express
iris_data = px.data.iris

# Create a scatter matrix plot with color mapping based on the "species" column
scatter_matrix = px.scatter_matrix(
    iris_data,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",  # Use the "species" column for color mapping
    title="Scatter Matrix Plot of Iris Dataset (Colored by Species)"
)

# Show the plot
scatter_matrix.show()
```

## Q5. What is Distplot? Using Plotly express, plot a distplot.

```python
A distplot, short for "distribution plot," is a graphical representation that combines a histogram and a kernel density
estimate (KDE) plot. It is used to visualize the distribution of a univariate dataset, allowing you to see the shape of
the data's distribution, its central tendency, and the presence of any peaks or modes. Distplots are particularly useful
for exploring the underlying distribution of continuous data.
```

```python
# Import the necessary libraries
import plotly.express as px

# Sample data (replace this with your own dataset)
data = [1.2, 1.5, 2.0, 2.2, 2.5, 2.5, 3.0, 3.2, 3.5, 3.5, 3.5, 4.0, 4.2, 4.5, 4.7]

# Create a distplot using Plotly Express
dist_plot = px.histogram(data, x=data, nbins=10, marginal="rug", title="Distplot Example")

# Show the plot
dist_plot.show()
```
